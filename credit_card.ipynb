```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
)
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\M.Nikhileswara\OneDrive\Desktop\Machine Learning\credit card fraud detection\creditcard.csv")
```

```python
df.head()
```

```python
df=df.drop(['Time'],axis=1)
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
df['Class'].value_counts()
```

```python
X=df.drop(['Class'],axis=1)
y=df['Class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
scaler=StandardScaler()
X_trainlr=scaler.fit_transform(X_train)
X_testlr=scaler.transform(X_test)
```

```python
lrmodel=LogisticRegression()
lrmodel.fit(X_trainlr,y_train)
```

```python
y_predlr=lrmodel.predict(X_testlr)
```

```python
print("test accuracy: ",accuracy_score(y_test,y_predlr))
print("precision : ",precision_score(y_test,y_predlr))
print("recall : ",recall_score(y_test,y_predlr))
print("f1 score: ",f1_score(y_test,y_predlr))
print("confusion matrix:\n",confusion_matrix(y_test,y_predlr))
```

### Balanced Model but recall is Low

## Oversampling

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE()
```

```python
X_over_train,y_over_train = smote.fit_resample(
    X_trainlr,
    y_train)
```

```python
X_over_train.shape
```

```python
y_train.value_counts()
```

```python
y_over_train.value_counts()
```

```python
lrmodel.fit(X_over_train,y_over_train)
```

```python
y_over_pred=lrmodel.predict(X_testlr)
```

```python
print("test accuracy: ",accuracy_score(y_test,y_over_pred))
print("precision : ",precision_score(y_test,y_over_pred))
print("recall : ",recall_score(y_test,y_over_pred))
print("f1 score: ",f1_score(y_test,y_over_pred))
print("confusion matrix:\n",confusion_matrix(y_test,y_over_pred))
```

### Very High recall but too many False Positives
